# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
plt.show()
```

## 在一张图形上多次绘图

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.show()
```

## 将图形保存为文件

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.show()
# plt.savefig('temp-fig\\MyPlot.png', format='png')
```

## 调整坐标轴

```python
axes = plt.axes()
axes.set_xlim([-5, 5])
axes.set_ylim([0, 1])
axes.set_xticks(list(range(-5, 6, 1)))   # range函数只支持int类型数据生成
axes.set_yticks(np.linspace(0, 1.0, 11)) # flot类型的range函数

plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.show()
```

## 添加网格

```python
axes = plt.axes()
axes.set_xlim([-5, 5])
axes.set_ylim([0, 1])
axes.set_xticks(list(range(-5, 6, 1)))   # range函数只支持int类型数据生成
axes.set_yticks(np.linspace(0, 1.0, 11)) # flot类型的range函数
axes.grid() # 添加网格
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.show()
```

## 修改线型和颜色

```python
axes = plt.axes()
axes.set_xlim([-5, 5])
axes.set_ylim([0, 1])
axes.set_xticks(list(range(-5, 6, 1)))   # range函数只支持int类型数据生成
axes.set_yticks(np.linspace(0, 1.0, 11)) # flot类型的range函数
axes.grid()
"""plot方法支持修改颜色和线型
颜色系：（可以单独使用）
'b' - 蓝色
'r' - 红色
'g' - 绿色

线型系：（跟颜色一起使用）
'-' - 实线
':' - 虚线
'--' - 短划线
'-.' 短划线和点的组合
"""
plt.plot(x, norm.pdf(x), 'b-') # 'b-'蓝色实线
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r:') # 'r:'红色虚线
plt.show()
```

## 标记坐标轴并添加图例

```python
axes = plt.axes()
axes.set_xlim([-5, 5])
axes.set_ylim([0, 1])
axes.set_xticks(list(range(-5, 6, 1)))   # range函数只支持int类型数据生成
axes.set_yticks(np.linspace(0, 1.0, 11)) # flot类型的range函数
axes.grid()
plt.plot(x, norm.pdf(x), 'b-') # 'b-'蓝色实线
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r:') # 'r:'红色虚线

"""标记坐标轴并添加图例
图例：
- loc：1 - 右上，2 - 左上，3 - 左下，4 - 右下
"""
plt.xlabel('Greebles')
plt.ylabel('Probability')
plt.legend(['Sneetches', 'Gacks'], loc=4)

plt.show()
```

## XKCD风格的图形

```python
plt.xkcd()

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none') # 去掉右边框
ax.spines['top'].set_color('none') # 去掉上边框
plt.xticks([])
plt.yticks([])
ax.set_ylim([-30, 10])

data = np.ones(100)
data[70:] -= np.arange(30)

"""添加注释
参数含义：注释内容，箭头指向位置，箭头配置，注释位置（左下角）
"""
plt.annotate(
    'THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',
    xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(15, -10))

plt.plot(data)

plt.xlabel('time')
plt.ylabel('my overall health')

plt.rcdefaults() # 回归正常模式
```

## 饼图

```python
"""生成饼图
只有values参数是必需的.
colors - 如果不指定，plt将使用默认的颜色配置
"""
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
explode = [0, 0, 0.2, 0, 0]
labels = ['India', 'United States', 'Russia', 'China', 'Europe']
plt.pie(values, colors=colors, labels=labels, explode=explode)
plt.title('Student Locations')
```

## 条形图

```python
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
plt.bar(range(0, 5), values, color=colors) # 参数color跟pie图的colors参数命名不一致
plt.show()
```

## 箱线图

```python
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100

data = np.concatenate((uniformSkewed, high_outliers, low_outliers)) # 连接数组

plt.boxplot(data)
plt.show()
```

## 查看函数的形状

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.arange(-10, 10, 0.001)

def sigmoid(z):
    return 1 / (1 + pow(math.e, -z))

axes = plt.axes()
axes.grid()
plt.plot(x, sigmoid(x))
plt.show()
```

```python
x = np.arange(0.01, 10, 0.001)
func = np.vectorize(math.log)
axes = plt.axes()
axes.grid()
plt.plot(x, func(x))
plt.show()
```
